```python
from transformers import BertTokenizer, BertForMaskedLM
import torch

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForMaskedLM.from_pretrained('bert-base-uncased')
```

```python
inputs = tokenizer(text, return_tensors='pt', max_length=512, truncation=True, padding='max_length')
```

```python
inputs['labels'] = inputs.input_ids.detach().clone()
```

```python
# create random array of floats with equal dimensions to input_ids tensor
rand = torch.rand(inputs.input_ids.shape)
# create mask array
mask_arr = (rand < 0.15) * (inputs.input_ids != 101) * \
           (inputs.input_ids != 102) * (inputs.input_ids != 0)
```

```python
selection = []

for i in range(inputs.input_ids.shape[0]):
    selection.append(
        torch.flatten(mask_arr[i].nonzero()).tolist()
    )
```

```python
for i in range(inputs.input_ids.shape[0]):
    inputs.input_ids[i, selection[i]] = 103
```

```python
class MeditationsDataset(torch.utils.data.Dataset):
    def __init__(self, encodings):
        self.encodings = encodings
    def __getitem__(self, idx):
        return {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
    def __len__(self):
        return len(self.encodings.input_ids)
```

```python
dataset = MeditationsDataset(inputs)
```

```python
loader = torch.utils.data.DataLoader(dataset, batch_size=16, shuffle=True)
```

```python
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
# and move our model over to the selected device
model.to(device)
```

```python
from transformers import AdamW

# activate training mode
model.train()
# initialize optimizer
optim = AdamW(model.parameters(), lr=5e-5)
```

```python
import gc
import torch
from tqdm import tqdm  # for our progress bar
import torch
gc.collect()
torch.cuda.empty_cache()

# torch.cuda.memory_summary(device=None, abbreviated=False)

epochs = 5

for epoch in range(epochs):
    # setup loop with TQDM and dataloader
    loop = tqdm(loader, leave=True)
    for batch in loop:
        # initialize calculated gradients (from prev step)
        optim.zero_grad()
        # pull all tensor batches required for training
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)
        # process
        outputs = model(input_ids, attention_mask=attention_mask,
                        labels=labels)
        # extract loss
        loss = outputs.loss
        # calculate loss for every parameter that needs grad update
        loss.backward()
        # update parameters
        optim.step()
        # print relevant info to progress bar
        loop.set_description(f'Epoch {epoch}')
        loop.set_postfix(loss=loss.item())
```

```python
import re
def mask_gender(string):
  """
  Change string's pronoun to that corresponding to a user given gender
  """
  regex_extracting_profession = r"[\s\w]*(\[[\w\s]*\])[\w\s]*(\[his\]|\[her\]|\[he\]|\[she\]|\[him\]|)"
  profession, gender = re.findall(regex_extracting_profession, string)[0]
  term_a = r'(\[his\])'
  term_b = r'(\[he\])'
  term_c = r'(\[him\])'
  term_d = r'(\[her\])'
  term_e = r'(\[she\])'
  if gender == '[he]' or  gender == '[him]' or  gender == '[his]':
    string = re.sub(term_a, '[MASK]', string)
    string = re.sub(term_c ,'[MASK]', string)
    string = re.sub(term_b, '[MASK]', string)
    # string = re.sub(term_c, '[him]', string)
  
  elif gender == '[she]' or gender == '[her]':
    # string = re.sub(term_a, '[her]', string)
    
    string = re.sub(term_d, '[MASK]', string)
    # string = re.sub(term_a, '[his]', string)
    string = re.sub(term_e ,'[MASK]]', string)
    # string = re.sub(term_c, '[him]', string)

  return string
```
